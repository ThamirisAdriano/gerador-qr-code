Alterar url da célula abaixo | sugestão de site com imagens em 3D: https://sketchfab.com/3d-models

```python
# URL do modelo 3D (substitua pelo link correto)
modelo_3d_url = "https://sketchfab.com/3d-models/cromossomo-c015eff83fd7456ea5a2cd7d87c2e7a9"
```

## ⚠️ Não edite esta célula abaixo! Ela é essencial para o funcionamento do notebook.


```python

# Gerar o QR Code
qr = qrcode.QRCode(
    version=1,
    error_correction=qrcode.constants.ERROR_CORRECT_L,
    box_size=10,
    border=4,
)
qr.add_data(modelo_3d_url)
qr.make(fit=True)

# Criar a imagem do QR Code
img = qr.make_image(fill="black", back_color="white")
```

Alterar o nome da imagem

```python
# Salvar a imagem
img.save("nome_da_imagem.png")

# Exibir a imagem
img.show()
```
